```python
#Part 1 : Process AutomationQ12: Data Cleaning & Normalization
# Q1: Create a file with 1000 lines of random strings

import random, string

with open("q1_1000_lines.txt", "w") as f:
    for _ in range(1000):
        line = ''.join(random.choices(string.ascii_letters + string.digits, k=50))
        f.write(line + "\n")

print("‚úÖ Q1: File with 1000 random lines created.")
```

```python
# Q2: Create one file of ~5MB with random strings

target_size = 5 * 1024 * 1024  # 5MB
current_size = 0

with open("q2_5mb_file.txt", "w") as f:
    while current_size < target_size:
        line = ''.join(random.choices(string.ascii_letters, k=100)) + "\n"
        f.write(line)
        current_size += len(line.encode())

print("‚úÖ Q2: File of approx 5MB created.")
```

```python
#Q3: Create 10 files each ~5MB with random strings

for i in range(1, 11):
    current_size = 0
    with open(f"q3_file_{i}.txt", "w") as f:
        while current_size < 5 * 1024 * 1024:
            line = ''.join(random.choices(string.ascii_letters, k=100)) + "\n"
            f.write(line)
            current_size += len(line.encode())

print("‚úÖ Q3: 10 files of ~5MB each created.")
```

```python
# Q4: Create 5 files of 1GB to 5GB with random strings

sizes_gb = [1, 2, 3, 4, 5]
for size in sizes_gb:
    current_size = 0
    with open(f"q4_file_{size}GB.txt", "w") as f:
        while current_size < size * 1024 * 1024:
            line = ''.join(random.choices(string.ascii_letters, k=100)) + "\n"
            f.write(line)
            current_size += len(line.encode())

print("‚úÖ Q4: Files of 1GB to 5GB simulated (as MB here for demo).")
```

```python
#Q5: Convert Q4 files to uppercase one by one

for size in sizes_gb:
    in_file = f"q4_file_{size}GB.txt"
    out_file = f"q5_file_{size}GB_upper.txt"
    with open(in_file, "r") as infile, open(out_file, "w") as outfile:
        for line in infile:
            outfile.write(line.upper())

print("‚úÖ Q5: Converted Q4 files to uppercase one by one.")
```

```python
#Q6: Convert Q4 files to uppercase using multithreading

import threading

def convert_to_upper_thread(filename):
    with open(filename, "r") as infile:
        data = infile.read().upper()
    with open("threaded_" + filename, "w") as outfile:
        outfile.write(data)

threads = []
for size in sizes_gb:
    file = f"q4_file_{size}GB.txt"
    t = threading.Thread(target=convert_to_upper_thread, args=(file,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("‚úÖ Q6: Converted Q4 files to uppercase using multi-threading.")
```

```python
#Q8: Download 10 Machine Learning videos from YouTube

!pip install yt-dlp

import os
import random
import subprocess

# Step 2: Set your search query
search_query = "Machine Learning"

# Step 3: Use yt-dlp to get top 20 video URLs (we'll pick random 10 from them)
search_command = f"yt-dlp 'ytsearch20:{search_query}' --get-id --quiet"
video_ids = subprocess.check_output(search_command, shell=True).decode().strip().split('\n')

# Step 4: Pick 10 random video IDs
selected_ids = random.sample(video_ids, 10)

# Step 5: Download videos
for i, vid_id in enumerate(selected_ids, 1):
    video_url = f"https://www.youtube.com/watch?v={vid_id}"
    print(f"‚ñ∂Ô∏è Downloading: {video_url}")
    subprocess.run([
        "yt-dlp", "-f", "bestvideo[ext=mp4]+bestaudio[ext=m4a]/mp4",
        "-o", f"q8_video_{i}.mp4", video_url
    ])
    print(f"‚úÖ Saved as: q8_video_{i}.mp4")

print("‚úÖ Q8: All videos downloaded successfully.")



```

```python
# Q9: Convert Downloaded YouTube Videos to Audio (.mp3)

from moviepy.editor import VideoFileClip
import os

# Convert 10 downloaded .mp4 videos to .mp3
for i in range(1, 11):
    video_file = f"q8_video_{i}.mp4"
    audio_file = f"q9_audio_{i}.mp3"
    if os.path.exists(video_file):
        try:
            clip = VideoFileClip(video_file)
            clip.audio.write_audiofile(audio_file)
            print(f"‚úÖ Converted {video_file} to {audio_file}")
        except Exception as e:
            print(f"‚ùå Failed to convert {video_file}: {e}")
    else:
        print(f"‚ö†Ô∏è Skipped: {video_file} not found")

print("‚úÖ Q9: All available videos converted to MP3.")
```

```python
#Q10: Store Video Names and Durations in a CSV File

import csv

# Prepare CSV file
with open("q10_video_data.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Video Name", "Duration (Seconds)"])

    for i in range(1, 11):
        video_file = f"q8_video_{i}.mp4"
        if os.path.exists(video_file):
            try:
                clip = VideoFileClip(video_file)
                duration = int(clip.duration)
                writer.writerow([video_file, duration])
                print(f"‚úÖ Stored: {video_file} with duration {duration}s")
            except Exception as e:
                print(f"‚ùå Error processing {video_file}: {e}")
        else:
            print(f"‚ö†Ô∏è Skipped: {video_file} not found")

print("‚úÖ Q10: CSV file created as 'q10_video_data.csv'.")
```

```python
#Part 2 : Data Analytics
#Q12: Data Cleaning & Normalization

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Step 1: Create a random dataset of 100 rows and 30 columns between [1,200]
np.random.seed(0)
df_q12 = pd.DataFrame(np.random.randint(1, 201, size=(100, 30)))

# (i) Replace values in [10, 60] with NA and count rows with missing values
df_q12_masked = df_q12.copy()
df_q12_masked = df_q12_masked.applymap(lambda x: np.nan if 10 <= x <= 60 else x)
missing_rows_count = df_q12_masked.isnull().any(axis=1).sum()
print("‚úÖ Q12(i): Rows with missing values =", missing_rows_count)

# (ii) Replace NA with column average
df_q12_filled = df_q12_masked.fillna(df_q12_masked.mean())
print("‚úÖ Q12(ii): NA values replaced with column means.")

# (iii) Pearson correlation heatmap and list of columns with any correlation ‚â§ 0.7
cor_matrix = df_q12_filled.corr()
sns.heatmap(cor_matrix, cmap='coolwarm', annot=False)
plt.title("Q12(iii): Pearson Correlation Heatmap")
plt.show()

# Select columns where any correlation value (with another) ‚â§ 0.7 (excluding self-correlation)
low_corr_columns = cor_matrix.columns[(cor_matrix < 0.7).any()]
print("‚úÖ Q12(iii): Columns with any correlation ‚â§ 0.7:\n", low_corr_columns.tolist())

# (iv) Normalize all values in the dataset between 0 and 10
scaler = MinMaxScaler(feature_range=(0, 10))
df_q12_normalized = pd.DataFrame(scaler.fit_transform(df_q12_filled), columns=df_q12.columns)
print("‚úÖ Q12(iv): Normalized all values between 0 and 10.")

# (v) Replace all values: if value ‚â§ 0.5 ‚Üí 1, else 0
df_q12_binary = df_q12_normalized.applymap(lambda x: 1 if x <= 0.5 else 0)
print("‚úÖ Q12(v): Values replaced with 1 if ‚â§ 0.5, else 0.")

```

```python
#Q13: Clustering with K-Means and Hierarchical Methods

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage

# Step 1: Create dataset of 500 rows and 10 columns with defined ranges
np.random.seed(0)
data_q13 = pd.DataFrame({
    **{f"C{i+1}": np.random.randint(-10, 11, 500) for i in range(4)},    # Columns 1-4
    **{f"C{i+5}": np.random.randint(10, 21, 500) for i in range(4)},     # Columns 5-8
    **{f"C{i+9}": np.random.randint(-100, 101, 500) for i in range(2)}   # Columns 9-10
})

# Step 2: Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_q13)

# -------------------- KMeans Clustering --------------------

# Use Elbow method to find optimal number of clusters
inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(data_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(K, inertia, marker='o')
plt.title('Q13(i): Elbow Method for Optimal k (K-Means)')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

# Apply KMeans with optimal k (assume 3 for simplicity here)
kmeans = KMeans(n_clusters=3, random_state=0)
labels_kmeans = kmeans.fit_predict(data_scaled)
data_q13['KMeans_Cluster'] = labels_kmeans
print("‚úÖ Q13(i): KMeans Clustering applied. Cluster labels added.")

# -------------------- Hierarchical Clustering --------------------

# Plot dendrogram for hierarchical clustering
linked = linkage(data_scaled, method='ward')
plt.figure(figsize=(10, 5))
dendrogram(linked, truncate_mode='lastp', p=20)
plt.title('Q13(ii): Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index or Cluster Size')
plt.ylabel('Distance')
plt.grid(True)
plt.show()

print("‚úÖ Q13(ii): Hierarchical Clustering dendrogram plotted.")
```

```python
# Q15: Statistical Testing on a Random Dataset

import pandas as pd
import numpy as np
from scipy import stats

# Create dataset
np.random.seed(15)
df_q15 = pd.DataFrame(np.random.uniform(5, 10, size=(500, 5)), columns=[f"C{i+1}" for i in range(5)])

# -------------------- (i) One-Sample t-Test --------------------
print("Q15(i): One-Sample t-Test (H0: mean = 7.5)\n")

for col in df_q15.columns:
    t_stat, p_val = stats.ttest_1samp(df_q15[col], popmean=7.5)
    print(f"{col}: t-statistic = {t_stat:.4f}, p-value = {p_val:.4f}")

# -------------------- (ii) Wilcoxon Signed-Rank Test --------------------
print("\nQ15(ii): Wilcoxon Signed-Rank Test (H0: median = 7.5)\n")

for col in df_q15.columns:
    stat, p_val = stats.wilcoxon(df_q15[col] - 7.5)
    print(f"{col}: stat = {stat:.4f}, p-value = {p_val:.4f}")

# -------------------- (iii) Two-Sample t-Test and Wilcoxon Rank-Sum Test Between C3 and C4 --------------------
print("\nQ15(iii): Two-Sample t-Test and Wilcoxon Rank-Sum Test Between C3 and C4\n")

# Two-sample t-Test (independent samples)
t_stat, p_val = stats.ttest_ind(df_q15["C3"], df_q15["C4"])
print(f"Two-sample t-Test (C3 vs C4): t-statistic = {t_stat:.4f}, p-value = {p_val:.4f}")

# Wilcoxon Rank-Sum Test (Mann-Whitney U Test for independent samples)
u_stat, p_val = stats.mannwhitneyu(df_q15["C3"], df_q15["C4"])
print(f"Wilcoxon Rank-Sum Test (C3 vs C4): U-statistic = {u_stat:.4f}, p-value = {p_val:.4f}")
```
